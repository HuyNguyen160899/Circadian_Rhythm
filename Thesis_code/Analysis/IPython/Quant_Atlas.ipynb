```python
import pandas as pd
```

```python
df = pd.read_csv('./01_Atlas_data/GSE54651_WFat_quant.txt', sep = '\t')
before = pd.read_csv('./01_Atlas_data/High_exp_WFat.csv')
```

```python
import mygene
mg = mygene.MyGeneInfo()
ens = df[df.columns[0]]
res = mg.querymany(ens, as_dataframe = True)
```

```python
pd.set_option('display.max_rows', 100)
df['Gene.ID'] = res[res.columns[4]].values
```

```python
df2 = df.dropna(subset=['Gene.ID']).reset_index(drop=True)
```

```python
avg = []
for i in range(len(df2)):
    average = sum(df2.iloc[i,1:].values)/8
    avg.append(average)
    
df_avg = pd.DataFrame(avg,columns=['Avg'])
df_avg.sort_values(by=['Avg'],inplace = True,ascending=False)
df_avg = df_avg[(df_avg['Avg'] >= 15)]
df_avg

final = df2.filter(items = df_avg.index.values, axis=0)
```

```python
start = list(final['Gene.ID'])
last = list(before['Gene.ID'])
start_idx = [x for x in start if x not in last]
len(start_idx)
```

```python
supp = final[final['Gene.ID'].isin(start_idx)]
supp
```

```python
supp.to_csv('WFat_supp.csv', index =None)
```
